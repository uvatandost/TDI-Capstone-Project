```python
# Columns to keep
cols = ['eventid', 'ccode' ,'countryname', 'startdate', 'enddate', 'duration','etype','actor1', 'target1', 'cgovtarget', 'rgovtarget',
       'issue1', 'repress']
latin_america1 = latin_america[cols]
africa1 = africa[cols]
```

```python
# Taking columns issue1 and etype for predictive analysis
print(africa.etype.value_counts())
print(africa.issue1.value_counts())
```

```python
# -9 category in etype and -99 category in issue1 represent unknown cases and since they represent less than %1 of data, 
# they are being dropped

africa = africa.loc[(africa.etype != -9) & (africa.issue1 != -99)]
```

```python
# Transforming the data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# onehot encoding event type(etype) and underlying issue (issue1) for latam
ohe_latam = OneHotEncoder().fit_transform(latin_america1[['etype','issue1']])

# onehot encoding event type(etype) and underlying issue (issue1) for Africa
ohe_africa = OneHotEncoder().fit_transform(africa1[['etype','issue1']])
```

```python
print(ohe_latam.toarray().shape)
print(ohe_africa.toarray().shape)
```

```python
# Defining new column names base on categories of one hot encoded data
col_names = ['Organized Demonstration', 'Spontaneous Demonstration', 'Organized Violent Riot', 'Spontaneous Violent Riot',
             'General Strike', 'Limited Strike', 'Pro-Government Violence (Repression)', 'Anti-Government Violence',
             'Extra-government Violence', 'Intra-government Violence','elections','economy, jobs', 
             'food, water, subsistence', 'environmental degradation','ethnic discrimination, ethnic issues',
             'religious discrimination, religious issues','education','foreign affairs/relations', 
             'domestic war, violence, terrorism', 'human rights, democracy', 'pro-government', 'economic resources/assets',
             'other','unknown, not-specified']
```

```python
df_latam = pd.DataFrame(ohe_latam.toarray())
df_latam.columns = col_names
```

```python
df_africa = pd.DataFrame(ohe_africa.toarray())
df_africa.columns = col_names
```

```python
# Combined and transformed dataset
df_full = pd.concat([df_latam, df_africa], ignore_index=True, sort=False)
df_full
```

```python
# Defining events that were peaceful
peaceful = df_full['Spontaneous Demonstration']+df_full['Organized Demonstration']+df_full['General Strike']+df_full['Limited Strike']
peaceful
```

```python
# Defining violent events
import numpy as np
violent = []
for row in peaceful:
    if row == 1:
        violent.append(0)
    else:
        violent.append(1)
print(len(violent))
print(np.sum(violent))
```

```python
# new dataset with target variable added
df_full1 = df_full[['elections','economy, jobs', 
             'food, water, subsistence', 'environmental degradation','ethnic discrimination, ethnic issues',
             'religious discrimination, religious issues','education','foreign affairs/relations', 
             'domestic war, violence, terrorism', 'human rights, democracy', 'pro-government', 'economic resources/assets',
             'other','unknown, not-specified']]
df_full1['violent'] = violent
df_full1
```

```python
# Downsampling the larger class to have two equal size classes for predictive model
df_full1_peaceful = df_full1[df_full1.violent==0]
df_full1_violent = df_full1[df_full1.violent==1].sample(7889)
df_full1_violent
```

```python
# Merging resampled and data
df_full11 = pd.merge(df_full1_peaceful, df_full1_violent, how='outer')
df_full11
```

```python
# Splitting training and test data
from sklearn.model_selection import train_test_split
X = df_full11.drop('violent', axis=1)
y = df_full11.violent
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=2)
```

```python
# Training the model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Predictions
model.predict(X_test)
```

```python
# Model accuracy
model.score(X_test, y_test)
```

```python
model.predict_proba(X_test)
```

```python
# Predicting probabilities
probs = model.predict_proba([[0,1,0,0,1,0,0,0,1,0,0,0,1,0]])
probs
```

```python
prob_nonviol, prob_viol = probs[0,0], probs[0,1]
prob_nonviol, prob_viol
```
